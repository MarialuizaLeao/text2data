```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_json('artists/artists.json', orient = 'index').reset_index() 
data.head()
```

```python
d = data.copy()
d.rename(columns = {'index': 'ID'}, inplace = True)
del (data)
d.head()
```

```python
d = d.dropna(axis = 0, subset = 'birthdate') 
d = d[d['decades'].map(lambda i: len(i)) > 0]
d['birthdate'] = d['birthdate'].str.replace('ca ', '')
d['birthdate'] = d['birthdate'].str.replace('ca. ', '')

d['birthdate'] = pd.to_datetime(d['birthdate'], yearfirst = True, errors = 'coerce')
d = d.dropna(axis = 0, subset = 'birthdate') 
```

```python
d.head()
```

```python
dsort = d.sort_values(by='birthdate')
dsort.reset_index(drop=True, inplace=True)
```

```python
dsort
```

```python
beatles = dsort[dsort['name']=='The Beatles']
beatles
```

```python
adjacency = np.zeros(shape = (len(dsort), len(dsort)))
# adjacency = np.zeros(shape = (1000,1000))

for i in range(len(dsort) - 1): 
    for influencer in dsort.loc[i, 'influencer']:
        id = "mn" + influencer.split('-mn')[1]
        
        if len(dsort[dsort['ID'] == id].index) == 1:
            id = dsort[dsort['ID'] == id].index[0]
            adjacency[i, id] = 1
    for follower in dsort.loc[i, 'followers']:
            id = "mn" + follower.split('-mn')[1]
            
            if len(dsort[dsort['ID'] == id].index) == 1:
                id = dsort[dsort['ID'] == id].index[0]
                adjacency[id, i] = 1
```
