```python
import requests
from bs4 import BeautifulSoup


page = requests.get('https://en.wikipedia.org/wiki/Immanuel_Kant')

soup = BeautifulSoup(page.content, "html.parser")

results = soup.find(id="bodyContent")

print(results.prettify())
```

```python
jobs = results.find_all("td", class_="infobox-full-data")

for job_element in jobs:
    print(job_element.prettify(), end="\n"*2)
```

```python
def check_whether_philosophy(name, URL):
    try:
        response = requests.get(URL)
    except:
        return 'LinkFailed'
    
    url_text = BeautifulSoup(response.text, 'html.parser')

    pp = url_text.find("div", class_="shortdescription nomobile noexcerpt noprint searchaux")
    
    if 'philosopher' in pp.text:
         return name
    else:
        return 'NotPhilosopher'
```

```python
influences = []
influenced = []

for job_element in jobs:
    
    if job_element.find_all(text = "Influences"):
        for element in job_element.find_all("a"):
            if element.get('title'):
                if element.get('href'):
                   influences.append(check_whether_philosophy(element.text, 'https://en.wikipedia.org/' + element.get('href')))
    
    if job_element.find_all(text = "Influenced"):
        for element in job_element.find_all("a"):
            if element.get('title'):
                if element.get('href'):
                   influenced.append(check_whether_philosophy(element.text, 'https://en.wikipedia.org/' + element.get('href')))    
```

```python
import re
# as per recommendation from @freylis, compile once only
CLEANR = re.compile('<.*?>') 

def cleanhtml(raw_html):
  cleantext = re.sub(CLEANR, '', raw_html)
  return cleantext
```

```python
URL = 'https://en.wikipedia.org/wiki/Aristotle'

response = requests.get(URL)

soup = BeautifulSoup(response.text, "html.parser")

for pag in soup.select('p'):
    print(cleanhtml(str(pag)))
```
